```python
import pandas as pd
```

```python
def alg(a, b, c, d, e, f, g, h):
    ka = 0.5
    kb = 0.5
    kc = 0.5
    kd = 0.5
    ke = 0.5
    kf = 1
    kg = 1
    kh = 1
    A = kd*d+ ke*e + kf*f + kh*h + kg*g
    B = ka*a
    C = ka*a
    D = kb*b
    E = kb*b
    F = kc*c
    G = kc*c
    H = kd*d + ke*e
    return A, B, C, D, E, F, G, H
```

```python
n = 1000
a, b, c, d, e, f, g, h = 1, 1, 1, 1, 1, 1, 1, 1
A = [0 for _ in range(n+1)] 
B = [0 for _ in range(n+1)] 
C = [0 for _ in range(n+1)] 
D = [0 for _ in range(n+1)] 
E = [0 for _ in range(n+1)] 
F = [0 for _ in range(n+1)] 
G = [0 for _ in range(n+1)] 
H = [0 for _ in range(n+1)]
A[0] = 1
B[0] = 1
C[0] = 1
D[0] = 1
E[0] = 1
F[0] = 1
G[0] = 1
H[0] = 1
for i in range(1, n+1):
    a, b, c, d, e, f, g, h = alg(a, b, c, d, e, f, g, h)
    A[i] = a
    B[i] = b
    C[i] = c
    D[i] = d
    E[i] = e
    F[i] = f
    G[i] = g
    H[i] = h
```

```python
df = pd.DataFrame({'A': A, 'B': B, 'C': C, 'D': D, 'E': E, 'F': F, 'G': G, 'H': H}).reset_index()
```
